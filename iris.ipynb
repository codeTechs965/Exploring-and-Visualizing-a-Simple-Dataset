```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_set=pd.read_csv("Iris.csv")
```

```python
data_set.head()
```

```python
data_set.info()
```

```python
data_set.describe()
```

```python
species_count=data_set["Species"].value_counts()
print(species_count)
```

```python
plt.scatter(species_count.index,species_count.values,color="Green",alpha=0.5)
plt.title("Species_Count")
plt.xlabel("Species")
plt.ylabel("Species_count")
plt.show()
```

```python
plt.hist(data_set["SepalWidthCm"])
plt.title("Frequency of Sepal Width Cm")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.boxplot(data_set["PetalLengthCm"])
plt.title("Box plot of Petal Length")
plt.ylabel("lengths")
plt.show()
```

```python
sns.histplot(data_set["PetalLengthCm"],color="red",bins=10)
plt.title("HistPlot of Sepal Length")
plt.show()
```
